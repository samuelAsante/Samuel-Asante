```python
#what is an outlier 
An outlier is an extremely high or extremely low data point relative to the nearest data point and the rest of the neighboring co-existing values in a data graph or dataset you're working with
```

```python
#Importance
Skewed Distributions:The presence of outliers can distort the shape and understanding of the data distribution. For instance, in a dataset containing water usage levels, a few outliers with unusually high consumption could create a false impression of higher water usage among the population than is the actual case.
```

```python
Data Visualization: Outliers can make data visualizations less informative or misleading. Removing or appropriately visualizing outliers can improve the clarity and interpretability of charts and plots
```

```python
Model Performance: Outliers can negatively impact the performance of machine learning models. Models that are sensitive to extreme values, like linear regression, can be heavily influenced by outliers. Removing or transforming outliers can lead to more accurate predictions.
```

```python
# Explain three common ways of detecting outliers in the given dataset
```

```python
Z-Score Method:To detect outliers using the Z-Score method, you would calculate the Z-Score for each data point in the "Potability" column. The Z-Score measures how many standard deviations a data point is away from the mean water usage. Data points with Z-Scores significantly higher or lower than the threshold are considered outliers in terms of water portability.
```

```python
Interquartile Range (IQR) Method:The IQR method focuses on the quartiles of the data distribution, specifically the first quartile (Q1) and the third quartile (Q3). Outliers are identified based on the range defined by the IQR. : Data points in the "Potability" column outside the IQR-defined range are treated as outliers.
```

```python
Visual Inspection (Box Plots and Scatter Plots):Visual methods involve creating plots to visually identify outliers. Box plots are particularly useful for this purpose. Data points that fall outside the "whiskers" of the box plot are considered outliers.
```
